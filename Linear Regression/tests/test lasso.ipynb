```python
import sys
import os
sys.path.append(os.path.abspath(".."))  # Adds parent directory to import path

from LassoRegressor import LassoRegressor  # Assuming the class is named like this
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 100
n_features = 20
n_informative = 5  # number of truly non-zero features

# True coefficients (only first `n_informative` are non-zero)
true_beta = np.zeros(n_features)
true_beta[:n_informative] = np.random.randn(n_informative)

# Create X and y
X = np.random.randn(n_samples, n_features)
y = X @ true_beta + np.random.randn(n_samples) * 0.5  # add noise

# Initiate model and fit
model = LassoRegressor()
model.fit(X, y, lam = 0.11)
beta_est = model.beta

# Plot
plt.figure(figsize=(10, 6))
plt.plot(true_beta, 'o-', label='True coefficients')
plt.plot(beta_est, 'x--', label='Lasso estimates')
plt.axhline(0, color='gray', linestyle=':')
plt.title("Lasso Feature Selection with Soft Thresholding")
plt.xlabel("Feature index")
plt.ylabel("Coefficient value")
plt.legend()
plt.grid(True)
plt.show()

# Print non-zero features
print("Non-zero features in Lasso estimate:")
print(np.nonzero(beta_est)[0])
```
